```python
# 1.- Crea un programa que reciba un rango de valores y almacene en una lista aquellos que sean múltiplos de 3.
print("Este programa encuentra todos los multiplos de 3.")
print("En un rango del 1 hasta el que numero elegido por el usuario.")
respuesta="si" # Asignar valor a "respuesta" con el finde repetir el Bucle, siempre cuando el usuario ingrese "si".
i=0
while respuesta=="si":
  limite=int(input("Escribe el limite superior: ")) # Designa el ultimo numero del rango, comenzando con 1.
  print("Los multiplos de 3 en un rango del 1 al ",limite,"son: ")
  for i in range(1,limite+1):
    if i%3==0: # Si "i"/3 da residuo 0
      multiplo=i # Asignar "i" en "multiplo"
      print("Tu multiplo es: "+str(multiplo)) # Transformar variable "multiplo" a str y despues imprimir "multiplo".
  respuesta=input("¿Quieres hacer otra suma (si/no)?")
  clear_output() # Borra la salida de una celda
print("Gracias por usar el programa.")
print("Vuelve pronto:)")
```
