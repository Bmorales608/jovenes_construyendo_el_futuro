```python
print("Este programa multiplica dos numeros")
str_num1=input("Escribe el primer numero: ") #Funcion "input" siempre arroja un "string"
str_num2=input("Escribe el segundo numero: ")
int_num1=int(str_num1)
int_num2=int(str_num2)
resultado=int_num1*int_num2
print("El resultado de",int_num1,"x",int_num2,"es",resultado)
#No puedes concatenar textos(str) con numeros(int)
```

```python
print("El resultado de {} x {} es {}".format(int_num1,int_num2,resultado)) # ".format() -> "
```

```python
print("Esta es una calculadora")
print("Elige el tipo de operacion del menu")
print("1. Sumar")
print("2. Restar")
print("3. Multiplicar")
print("4. Dividir")
res=input("¿Que quieres hacer? " )
num1=float(input("Escribe el primer numero: "))
num2=float(input("Escribe el segundo numero: "))
if res=="sumar":
  print("{} + {} = {}" .format(num1,num2,num1+num2))
elif res=="restar":
  print("{} - {} = {}" .format(num1,num2,num1-num2))
elif res=="multiplicar":
  print("{} * {} = {}" .format(num1,num2,num1*num2))
elif res=="dividir":
  print("{} / {} = {}" .format(num1,num2,num1/num2))
else:
  print("Comando no valido")  
```

```python
num1=10
num2=10
if num1!=num2:
  print("No son iguales")
elif num1>=num2: #Se ejecuta la primera respuesta que sea acorde a la pregunta, ignorando el resto de respuestas.
  print("El primero es mayor o igual")
elif num1==num2:
  print("Son iguales")
else:
  print("Ninguna condicion")
```
